# creating age profile

```python
# If the user is younger than 26, then user will be labeled as “Young adult”.
   #If the user is 26 and younger than 65, they will be labeled a “Adult”.
      #Users 65 and older will be labeled as "Retired".
```

```python
#If the user is younger than 26, then user will be labeled as “Young adult”
merged_df.loc[merged_df['age'] < 26, 'age_profile'] = 'Young adult'
```

```python
#If the user is 26 and younger than 65, they will be labeled a “Adult”
merged_df.loc[(merged_df['age'] < 65) & (merged_df['age'] >= 26), 'age_profile'] = 'Adult' 
```

```python
#Users 65 and older will be labeled as "Retired"
merged_df.loc[merged_df['age'] >= 65, 'age_profile'] = 'Retired'
```

```python
#Count values
merged_df['age_profile'].value_counts(dropna = False)
#Most users are adults age 26 and older and younger than 65 years old.
```

```python
#Check flags
merged_df.groupby('age_profile').agg({'age' : ['min', 'max']})
#nothing odd on the results
```

```python
#Evaluate "number_of_dependents"
merged_df['number_of_dependents'].describe()
#User have non or up to 3 kids
```

```python
#Printing the frequency of a column will quickly inform you which values appear more often within that column.
merged_df['number_of_dependents'].value_counts(dropna = False)
#Most users have at least 1 kid 
```

```python
#The folllowing conditions are applied:
#If the user does not have dependents (=0), then user will be labeled as “No kids”.
#If the user has 1 or more dependents then, they will be labeled as “With kids”.

result = []

for value in merged_df["number_of_dependents"]:
  if value == 0:
    result.append("No kids")
  else:
    result.append("With kids")
```

```python
result
```

```python
#Adding the results to your dataframe
merged_df['dependents_profile'] = result
```

```python
#See results
merged_df.head()
#variable "dependents_profile" was added
```

```python
#Counting values within new column
merged_df['dependents_profile'].value_counts(dropna = False)
#Nothing odd and matches with previous frequency
```

```python
#Check results
merged_df[['dependents_profile', 'number_of_dependents']].head()
#nothing odd on the results
```

# creating income profile

```python
merged_df['income'].describe()
```

```python
#The folllowing conditions are applied:
#Users with incomes of 60k per year or lower will be labeled as "Low income".
#Users with incomes higher than 60k but less than 250k will be labeled as "Middle income".
#User with incomes of 250k or higher will be labeled as "High income".

result = []

for value in merged_df["income"]:
  if value < 60000:
    result.append("Low income")
  elif value >= 250000:
    result.append("High income")
  else:
    result.append("Middle income")
```

```python
result
```

```python
#Adding the results to the df
merged_df['income_profile'] = result
```

# customer expenditure

```python
#Calculating the average price of products purchased per customer profile = ("mean_price_per_customer_profile")
merged_df['mean_price_per_customer_profile'] = merged_df.groupby(['customer_profile'])['prices'].transform(np.mean)
```

```python
#Calculating the max price of products purchased per customer profile = ("max_price_per_customer_profile")
merged_df['max_price_per_customer_profile'] = merged_df.groupby(['customer_profile'])['prices'].transform(np.max)
```

```python
#Calculating the min price of products purchased per customer profile = ("min_price_per_customer_profile")
merged_df['min_price_per_customer_profile'] = merged_df.groupby(['customer_profile'])['prices'].transform(np.min)
```

```python
#Check flags
merged_df.groupby('customer_profile').agg({'prices' : ['min', 'max', 'mean']})
```

```python
#See results
merged_df[['customer_profile', 'mean_price_per_customer_profile', 'max_price_per_customer_profile', 'min_price_per_customer_profile', 'prices']].head()
```

# customer application usage frequency

```python
#Calculating the mean usage frequency per customer profile = ("mean_usage_per_customer_profile")
merged_df['mean_usage_per_customer_profile'] = merged_df.groupby(['customer_profile'])['days_since_prior_order'].transform(np.mean)
```

```python
#Calculating the max usage frequency per customer profile = ("max_usage_per_customer_profile")
merged_df['max_usage_per_customer_profile'] =  merged_df.groupby(['customer_profile'])['days_since_prior_order'].transform(np.max)
```

```python
#Calculating the min usage frequency per customer profile = ("min_usage_per_customer_profile")
merged_df['min_usage_per_customer_profile'] =  merged_df.groupby(['customer_profile'])['days_since_prior_order'].transform(np.min)
```

```python
#Check flags
merged_df.groupby('customer_profile').agg({'days_since_prior_order' : ['min', 'max', 'mean']})
```

```python
#See results
merged_df[['customer_profile', 'mean_usage_per_customer_profile', 'max_usage_per_customer_profile', 'min_usage_per_customer_profile', 'days_since_prior_order']].head()
```
