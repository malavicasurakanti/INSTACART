frequencyof order between,non-customers, regular customer & frequcent customer

1.Non-frequent Customer: A customer is categorized as a "Non-frequent Customer" if the median value of their "days_since_prior_order" is greater than 20 days.

2.Regular Customer: A customer is classified as a "Regular Customer" if their "days_since_prior_order" median falls between 10 and 20 days, inclusive.

3.Frequent Customer: A customer is labeled a "Frequent Customer" if the median of their "days_since_prior_order" is 10 days or less.


```python
# Creating median_order_frequency column which show customers based on their median days_since_last_order
df['median_order_frequency'] = df.groupby(['user_id'])['days_since_prior_order'].transform('median')
```

```python
df.head()
```

```python
#A. Creating "order_frequency_flag" for median_orders > 20 as "Non-frequent customer"
df.loc[df['median_order_frequency'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
#B. Creating "order_frequency_flag" for median_orders <= 20 & >10 as "regular customer"
df.loc[(df['median_order_frequency'] <= 20) & (df['median_order_frequency'] > 10), 'order_frequency_flag'] = 'Regular customer'
```

```python
#C. Creating "order_frequency_flag" for median_orders <= 10 as "frequent customer"
df.loc[df['median_order_frequency'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
# checking "order_frequency_flag" frequency
df['order_frequency_flag'].value_counts(dropna = False)
```
