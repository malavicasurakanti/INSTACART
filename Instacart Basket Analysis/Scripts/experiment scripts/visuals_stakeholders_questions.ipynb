# observations from the above histogram
Skewness: The distribution is right-skewed, meaning there are a few items with much higher prices compared to the
    majority. The skewness indicates that while most items fall within a lower price range, there are outliers 
    or more expensive items.

Bins and Frequency: The histogram has a total of 15 bins. 
    The y-axis represents the frequency of items within each price range.
    The tallest bar indicates the price range with the highest frequency of items.

Range of Prices: Prices span from under 2 to over 14, with varying frequencies.

Outliers: There seem to be a few outliers on the higher end of the price range, 
    as indicated by the bins with very low frequency past the 10 mark on the x-axis.

Central Tendency: The majority of the data is clustered around the lower end of the price range. 
    There is a prominent peak, suggesting that most prices fall into a relatively low price category, 
    likely under $5 based on the scale.

```python
#creating a line chart to explain difference in expenditure (the â€œpricesâ€ column) depending on the hour of the day.

# Assuming 'df' is your DataFrame with 'order_hour_of_day' and 'prices' columns

# Step 1 & 2: Group by 'order_hour_of_day' and calculate the mean of 'prices'
hourly_expenditure = df_combined.groupby('order_hour_of_day')['prices'].mean().reset_index()

# Step 3: Sample your data if needed. For example, if you have a very large dataset:
# You can use DataFrame.sample() method to get a random sample
# sample_df = hourly_expenditure.sample(frac=0.3, random_state=1)

# Step 4: Plot a line chart
line_hour=plt.figure(figsize=(14, 7))  # Adjust the size of the figure as necessary
sns.lineplot(data=hourly_expenditure, x='order_hour_of_day', y='prices')

plt.title('Average Expenditure by Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Average Expenditure')
plt.xticks(range(0, 24))  # Assuming hours go from 0 to 23
plt.grid(True)  # Optional: Add grid for better readability

# Show the plot
plt.show()
```

```python

# Step 1: Group by 'age' and calculate the mean of 'number_of_dependents'
age_dependents = df_combined.groupby('age')['n_dependants'].mean().reset_index()

# Step 2: Plot a line chart
line_age=plt.figure(figsize=(14, 7))  # Adjust the size of the figure as necessary
sns.lineplot(data=age_dependents, x='age', y='n_dependants')

plt.title('Average Number of Dependents by Age')
plt.xlabel('Age')
plt.ylabel('Average Number of Dependents')
plt.grid(True)  # Optional: Add grid for better readability

# Show the plot
plt.show()
```
