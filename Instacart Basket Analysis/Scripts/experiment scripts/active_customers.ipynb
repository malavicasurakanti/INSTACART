```python
import pandas as pd

# Assuming df_combined is your DataFrame

# Step 1: Determine the number of orders per customer
order_counts = df_combined.groupby('user_id')['order_number'].max()

# Step 2: Create the exclusion flag
order_counts_df = order_counts.reset_index(name='total_orders')
order_counts_df['exclusion_flag'] = order_counts_df['total_orders'] < 5

# Step 3: Merge the flag back to the original DataFrame
df_combined_with_flag = pd.merge(df_combined, order_counts_df[['user_id', 'exclusion_flag']], on='user_id')

# Step 4: Create the conditional subset
active_customers_df = df_combined_with_flag[df_combined_with_flag['exclusion_flag'] == False]
```

```python
active_customers_df.head()
```

```python
# Assuming df_combined is your original DataFrame
# Assuming active_customers_df is the DataFrame after applying the exclusion flag

# Count unique customers in the original DataFrame
original_customer_count = df_combined['user_id'].nunique()

# Count unique customers in the subset DataFrame
active_customer_count = active_customers_df['user_id'].nunique()

# Calculate the number of customers eliminated
eliminated_customer_count = original_customer_count - active_customer_count

print(f"Number of eliminated customers: {eliminated_customer_count}")
```
