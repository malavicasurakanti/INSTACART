```python
# Change "order_dow" column name to "order_day_of_week."

df_combined.rename(columns = {'order-day_of_day' : 'orders_day_of_week'}, inplace = True)
```

```python
df_combined['orders_day_of_week'].value_counts(dropna = False)
```

```python
# Use for-loop to determine whether orders are on "busiest day" (0 = Saturday), "least busy" (4 = Wednesday), or "regularly busy" (other days of the week).

result = []

for value in df_combined["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
# Create new column in ords_prods_merge to view results in context.

df_combined['busiest_day'] = result
```

```python
df_combined['busiest_day'].value_counts(dropna = False)
```

```python
# Create column â€œbusiest_days" from "orders_day_of_week" to identify two busiest days (0, 1), slowest days (4, 3), or regularly busy (all other days).

result_2 = []

for value in df_combined["orders_day_of_week"]:
  if value == 0 or value == 1:
    result_2.append("Busiest days")
  elif value == 4 or value == 3:
    result_2.append("Slowest days")
  else:
    result_2.append("Regularly busy")
```

```python
result_2
```

```python
# Create new 'busiest_days' column in ords_prods_merge to view results in context.

df_combined['busiest_days'] = result_2
```

```python
df_combined['busiest_days'].value_counts(dropna = False)
```

```python
# Check value counts in "order_hour_of_day" column.

df_combined['order_hour_of_day'].value_counts()
```

```python
# Create for-loop if statement labeling periods of time as â€œMost orders,â€ â€œAverage orders,â€ and â€œFewest orders.â€

result_3 = []

for value in df_combined["order_hour_of_day"]:
  if value in [10, 11, 14, 15, 13, 12, 16, 9]:
    result_3.append("Most orders")
  elif value in [23, 6, 0, 1, 5, 2, 4, 3]:
    result_3.append("Fewest orders")
  else:
    result_3.append("Average orders")
```

```python
result_3
```

```python
# Create new column "busiest_period_of_day" in ords_prods_merge.

df_combined['busiest_period_of_day'] = result_3
```

```python
# Print value counts in "busiest_period_of_day" column.

df_combined['busiest_period_of_day'].value_counts(dropna = False)
```
